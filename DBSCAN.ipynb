```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn .cluster import DBSCAN
data=np.array([[500,10],
               [15000,100],
               [510,11],
               [520,12],
               [530,13],
               [20000,180]])
db=DBSCAN(eps=60,min_samples=2)
labels =db.fit_predict(data)
df=pd.DataFrame(data,columns=['Amount','Time'])
df['label']=labels
print("DBSCAN OUTPUT:\n")
print(df)
plt.figure(figsize=(8,5))
color=['green'if label==0 else 'red' if label==1 else 'blue' for label in labels]
plt.scatter(df['Amount'],df['Time'],c=color)
plt.xlabel('Amount')
plt.ylabel('Time')
plt.show()
```
